```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("data/Rwanda/actual_predicted.csv")

figure = plt.figure()
ax=figure.add_subplot(1,1,1)

p1=ax.scatter(data['cluster'],data['Actual_wealth'], marker="^",s=50, edgecolors='none', alpha=0.5)
p2=ax.scatter(data['cluster'],data['Predicted_wealth'],marker="o", s=80, edgecolors='none', alpha=1, facecolor="none", edgecolor="orange")

ax.legend([p1,p2],["Actual Wealth","Predicted Wealth"])

ax.set_title('Actual vs Predicted wealth index', fontsize=14)
ax.set_xlabel('Number of Clusters', fontsize=12)
ax.set_ylabel('Wealth index', fontsize=12)

plt.savefig('plot.png',dpi=1200)
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# set the filepath and load in a shapefile
fp = 'data\DHS\RWGE61FL'
fp1= 'Mozambique\MZGE7AFL1'
fp2= 'Madagascar\MDGE71FL'
fp3= 'Nigeria\_NGGE7AFL1'

mp ='data\shp\Sector_Boundary_2012\Sector_Boundary_2012.shp'
mp1='Mozambique\shp\moz_admbnda_adm1_ine_20190607.shp'
mp2='Madagascar\shp\Madagascar.shp'
mp3='Nigeria\\NIR-level_1.shp'

map_df = gpd.read_file(fp)
map_df1 = gpd.read_file(fp1)
map_df2 = gpd.read_file(fp2)
map_df3 = gpd.read_file(fp3)

map_mp = gpd.read_file(mp)
map_mp1 = gpd.read_file(mp1)
map_mp2 = gpd.read_file(mp2)
map_mp3 = gpd.read_file(mp3)

# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
print(map_df.columns)
print(map_df1.columns)
print(map_df2.columns)
print(map_df3.columns)

map_mp.plot()
map_mp1.plot()
map_mp2.plot()
map_mp3.plot()
```

```python
data_map = map_df.rename(index=str, columns={'DHSCLUST': 'cluster'})
data_map.head()
data_map1 = map_df1.rename(index=str, columns={'DHSCLUST': 'cluster'})
data_map1.head()
data_map2 = map_df2.rename(index=str, columns={'DHSCLUST': 'cluster'})
data_map2.head()
data_map3 = map_df3.rename(index=str, columns={'DHSCLUST': 'cluster'})
data_map3.head()

df = pd.read_csv('intermediate_files/rwanda_cluster_avg_asset_2010.csv', header=0)
df.head()
df1 = pd.read_csv('Mozambique/Mozambique_cluster_avg_asset_2010.csv', header=0)
df1.head()
df2 = pd.read_csv('Madagascar/Madagascar_cluster_avg_asset_2010.csv', header=0)
df2.head()
df3 = pd.read_csv('Nigeria//Nigeria_cluster_avg_asset_2010.csv', header=0)
df3.head()
df31 = pd.read_csv('New Data//Nigeria//Nigeria_cluster_avg_asset_2010.csv', header=0)
df31.head()
df21 = pd.read_csv('New Data/Madagascar/madagascar_cluster_avg_asset_2010.csv', header=0)
df21.head()
df01 = pd.read_csv('New Data/Rwanda/rwanda_cluster_avg_asset_2010.csv', header=0)
df01.head()
```

```python
merged = data_map.set_index('cluster').join(df.set_index('cluster'))
merged.head()
merged1 = data_map1.set_index('cluster').join(df1.set_index('cluster'))
merged1.head()
merged2 = data_map2.set_index('cluster').join(df2.set_index('cluster'))
merged2.head()
merged3 = data_map3.set_index('cluster').join(df3.set_index('cluster'))
merged3.head()
merged21 = data_map2.set_index('cluster').join(df21.set_index('cluster'))
merged21.head()
merged01 = data_map.set_index('cluster').join(df01.set_index('cluster'))
merged01.head()
merged31 = data_map3.set_index('cluster').join(df31.set_index('cluster'))
merged31.head()

merged21.plot()
map_mp2.plot()
```
